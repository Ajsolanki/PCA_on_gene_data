```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(df.transpose())
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```

```python
principalDf
```

```python
col_df['histological_type_other'].value_counts()
```

```python
neuro_type=['neuroendocrine carcinoma nos','neuroendocrine','82463 neuroendocrine carcinoma nos','neuroendocrine carcinoma']
        
A=col_df['histological_type_other'].values

for i, ele in enumerate(A):
    if ele in (neuro_type):
        A[i]='neuroendocrine'

principalDf['histological_type_other']=A
principalDf['histological_type_other'].value_counts()
```

It's clearly shown that there is only 8 sample is belongs to neuroendocrine type 

```python
sns.set(rc={'figure.figsize':(15,10)})
ax = sns.scatterplot(x="principal component 1", y="principal component 2", data=principalDf, hue="histological_type_other",s=80)
```

```python
pca.explained_variance_ratio_

print('The amount of varience explained by first 2 components are : ',sum(pca.explained_variance_ratio_)*100)
```
